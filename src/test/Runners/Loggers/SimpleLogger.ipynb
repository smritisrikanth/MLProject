```python
from checker_players.Players.RandomPlayer import RandomPlayer
from checker_players.Checker_Environment.Board import Board
from checker_players.Runners.Loggers.SimpleLogger import SimpleLogger 
```

```python
board = Board()
player1 = RandomPlayer(board.P1)
player2 = RandomPlayer(board.P2)
logger = SimpleLogger()
```

```python
logger.begin()
for i in range(1000):
    if board.is_game_over():
        break
    curr_player = player1 if board.player_turn else player2
    board.make_move(curr_player.play(board))
    logger.apply(board, player1, player2)
logger.end()
winner, current_turn_number, board_state_history, board_time_per_move_history = logger.exit()
```

```python
print(winner)
print(current_turn_number)
print(board_state_history)
print(board_time_per_move_history)
```

```python
for state in board_state_history:
    Board(state).print_board()
    print("***********************************************************")
```
